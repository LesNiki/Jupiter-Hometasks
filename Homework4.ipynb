## Задача 2

### 2.1 Изучите распределение домов от наличия вида на набережную

```python
data = df['waterfront'].value_counts()
data
```

```python
names = data.index
values = data.values

plt.figure(figsize=(6, 4))
plt.title('Распределение домов от наличия вида на набережную')
plt.pie(values, autopct='%.1f%%')
plt.legend(['no', 'yes']);
```

```python
# Видно, что вид на набережную имеют меньше 1% домов
```

### 2.2 Изучите распределение этажей домов

```python
data = df['floors'].value_counts()
data
```

```python
plt.figure(figsize=(6, 4))
sns.barplot(x = data.index, y = data.values)
plt.title('Распределение этажности домов')
plt.xlabel('Этажи')
plt.ylabel('Количество');
```

```python
# Дома преимущественно одно- и двухэтажные
```

### 2.3 Изучите распределение состояния домов

```python
data = df['condition'].value_counts()
data
```

```python
plt.figure(figsize=(8, 6))
plots = sns.barplot(x = data.index, y = data.values)

#добавила подписи к столбикам
for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.2f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=12, xytext=(0, 5),
                   textcoords='offset points')

plt.title('Распределение состояния домов')
plt.xlabel('Состояние')
plt.ylabel('Количество');
```

```python
# Преимущественное количество домов имеет оценку состояния 3
```
